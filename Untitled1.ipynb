### import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import warnings
warnings.filterwarnings('ignore')

# Настройка стиля графиков
plt.style.use('ggplot')
sns.set_palette("Set2")

# Подключение к БД
engine = create_engine('postgresql://postgres:mysecretpassword@localhost:5432/postgres')

# Загрузка таблиц
books = pd.read_sql('SELECT * FROM books', engine)
authors = pd.read_sql('SELECT * FROM authors', engine)
customers = pd.read_sql('SELECT * FROM customers', engine)
orders = pd.read_sql('SELECT * FROM orders', engine)
order_details = pd.read_sql('SELECT * FROM order_details', engine)

print("Размеры таблиц:")
print(f"books: {books.shape}")
print(f"authors: {authors.shape}")
print(f"customers: {customers.shape}")
print(f"orders: {orders.shape}")
print(f"order_details: {order_details.shape}")


```python
# Проверка пропущенных значений
print("Пропущенные значения:")
print(books.isnull().sum())
print(orders.isnull().sum())

# Статистика по числовым полям
print("\nСтатистика по книгам:")
print(books[['price', 'publication_year']].describe())

print("\nСтатистика по заказам:")
print(orders[['total_amount']].describe())
```

```python
# Объединяем order_details с books
book_sales = order_details.merge(books, on='book_id', suffixes=('_detail', '_book'))
book_sales['revenue'] = book_sales['quantity'] * book_sales['price_book']

# Группировка по книгам
top_books = book_sales.groupby('title').agg({
    'revenue': 'sum',
    'quantity': 'sum',
    'book_id': 'count'
}).rename(columns={'book_id': 'order_count'}).sort_values('revenue', ascending=False).head(10)

print(top_books)

# Визуализация
plt.figure(figsize=(10,6))
plt.barh(top_books.index, top_books['revenue'], color='skyblue')
plt.xlabel('Выручка')
plt.title('Топ-10 книг по выручке')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# RFM-анализ
current_date = orders['order_date'].max() + pd.Timedelta(days=1)

rfm = customers.merge(orders, on='customer_id', how='left').groupby('customer_id').agg({
    'order_date': lambda x: (current_date - x.max()).days if not x.isnull().all() else None,
    'order_id': 'count',
    'total_amount': 'sum'
}).rename(columns={'order_date': 'recency', 'order_id': 'frequency', 'total_amount': 'monetary'})

# Убираем клиентов без заказов
rfm = rfm.dropna()

# Присваиваем ранги (1-5)
rfm['r_score'] = pd.qcut(rfm['recency'], 5, labels=[5,4,3,2,1], duplicates='drop')
rfm['f_score'] = pd.qcut(rfm['frequency'], 5, labels=[1,2,3,4,5], duplicates='drop')
rfm['m_score'] = pd.qcut(rfm['monetary'], 5, labels=[1,2,3,4,5], duplicates='drop')

# Сегментация
def segment(row):
    if row['r_score'] == 5 and row['f_score'] == 5 and row['m_score'] == 5:
        return 'Чемпионы'
    elif row['r_score'] >= 4 and row['f_score'] >= 4 and row['m_score'] >= 4:
        return 'Лояльные'
    elif row['r_score'] >= 3 and row['f_score'] >= 3 and row['m_score'] >= 3:
        return 'Потенциальные'
    elif row['r_score'] <= 2 and row['f_score'] <= 2 and row['m_score'] <= 2:
        return 'Спящие'
    else:
        return 'Смешанный'

rfm['segment'] = rfm.apply(segment, axis=1)

# Визуализация сегментов
segment_counts = rfm['segment'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(segment_counts, labels=segment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Распределение клиентов по RFM-сегментам')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Подготовка данных по месяцам
orders['order_date'] = pd.to_datetime(orders['order_date'])
monthly = orders.set_index('order_date').resample('ME')['total_amount'].sum().reset_index()
monthly['month_num'] = range(len(monthly))

X = monthly[['month_num']]
y = monthly['total_amount']
model = LinearRegression().fit(X, y)
pred = model.predict([[len(monthly)]])
print(f"Прогноз на следующий месяц: {pred[0]:.2f}")
```

```python
# Преобразуем дату, если ещё не сделано
orders['order_date'] = pd.to_datetime(orders['order_date'])

# Добавляем название дня недели
orders['weekday'] = orders['order_date'].dt.day_name()

# Группировка
weekday_stats = orders.groupby('weekday').agg(
    orders_count=('order_id', 'count'),
    total_revenue=('total_amount', 'sum'),
    avg_order=('total_amount', 'mean')
).reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

print(weekday_stats)

# Визуализация
plt.figure(figsize=(10,5))
plt.bar(weekday_stats.index, weekday_stats['orders_count'], color='teal')
plt.title('Количество заказов по дням недели')
plt.xticks(rotation=45)
plt.ylabel('Количество заказов')
plt.tight_layout()
plt.show()
```

```python
orders['year_month'] = orders['order_date'].dt.to_period('M')
monthly_stats = orders.groupby('year_month').agg(
    orders_count=('order_id', 'count'),
    total_revenue=('total_amount', 'sum')
).reset_index()
monthly_stats['year_month'] = monthly_stats['year_month'].astype(str)

print(monthly_stats)

plt.figure(figsize=(12,5))
plt.plot(monthly_stats['year_month'], monthly_stats['total_revenue'], marker='o')
plt.title('Динамика выручки по месяцам')
plt.xticks(rotation=45)
plt.ylabel('Выручка')
plt.tight_layout()
plt.show()
```

```python
# Количество заказов на клиента
customer_orders = orders.groupby('customer_id').size().reset_index(name='order_count')

# Статистика
print("Распределение количества заказов на клиента:")
print(customer_orders['order_count'].value_counts().sort_index())

# Доля клиентов с повторными покупками (более 1 заказа)
repeat_customers = customer_orders[customer_orders['order_count'] > 1]
total_customers = len(customer_orders)
print(f"\nКлиентов с повторными заказами: {len(repeat_customers)} ({len(repeat_customers)/total_customers:.1%})")

# Доля выручки от повторных клиентов
# Сумма заказов по каждому клиенту
customer_revenue = orders.groupby('customer_id')['total_amount'].sum().reset_index()
customer_revenue = customer_revenue.merge(customer_orders, on='customer_id')
repeat_revenue = customer_revenue[customer_revenue['order_count'] > 1]['total_amount'].sum()
total_revenue = customer_revenue['total_amount'].sum()
print(f"Доля выручки от повторных клиентов: {repeat_revenue/total_revenue:.1%}")

# Визуализация распределения
plt.figure(figsize=(8,5))
customer_orders['order_count'].value_counts().sort_index().plot(kind='bar')
plt.title('Распределение клиентов по количеству заказов')
plt.xlabel('Количество заказов')
plt.ylabel('Количество клиентов')
plt.tight_layout()
plt.show()
```

```python
top_customers = customer_orders.sort_values('order_count', ascending=False).head(10)
print("Топ-10 клиентов по числу заказов:")
print(top_customers.merge(customers[['customer_id', 'name']], on='customer_id'))
```

```python
# Выбираем только заказы, в которых больше одной книги
multi_book_orders = order_details.groupby('order_id').filter(lambda x: len(x) > 1)

# Объединяем таблицу саму с собой по order_id, чтобы получить все пары книг в одном заказе
pairs = multi_book_orders.merge(
    multi_book_orders[['order_id', 'book_id']], 
    on='order_id', 
    suffixes=('_a', '_b')
)

# Оставляем только пары с разными книгами (чтобы избежать (a,a))
pairs = pairs[pairs['book_id_a'] < pairs['book_id_b']]

# Группируем по парам книг
pair_counts = pairs.groupby(['book_id_a', 'book_id_b']).size().reset_index(name='freq')

# Сортируем по частоте
top_pairs = pair_counts.sort_values('freq', ascending=False).head(10)

# Добавим названия для первой книги
top_pairs = top_pairs.merge(books[['book_id', 'title']], left_on='book_id_a', right_on='book_id', how='left')
top_pairs = top_pairs.rename(columns={'title': 'title_a'})
top_pairs = top_pairs.drop(columns=['book_id'])

# Добавим названия для второй книги
top_pairs = top_pairs.merge(books[['book_id', 'title']], left_on='book_id_b', right_on='book_id', how='left')
top_pairs = top_pairs.rename(columns={'title': 'title_b'})
top_pairs = top_pairs.drop(columns=['book_id'])

# Выбираем нужные колонки
top_pairs = top_pairs[['title_a', 'title_b', 'freq']]

print("Топ-10 пар книг, которые покупают вместе:")
print(top_pairs)
```

```python
# Определяем месяц первой покупки для каждого клиента
first_order = orders.groupby('customer_id')['order_date'].min().reset_index()
first_order['cohort_month'] = first_order['order_date'].dt.to_period('M')

# Добавляем информацию о когорте к каждому заказу
orders_with_cohort = orders.merge(first_order[['customer_id', 'cohort_month']], on='customer_id')
orders_with_cohort['order_month'] = orders_with_cohort['order_date'].dt.to_period('M')
orders_with_cohort['cohort_index'] = (orders_with_cohort['order_month'] - orders_with_cohort['cohort_month']).apply(lambda x: x.n)

# Строим когортную таблицу: количество уникальных клиентов в каждой когорте по месяцам
cohort_data = orders_with_cohort.groupby(['cohort_month', 'cohort_index'])['customer_id'].nunique().reset_index()
cohort_pivot = cohort_data.pivot(index='cohort_month', columns='cohort_index', values='customer_id')

print("Когортная таблица (количество активных клиентов):")
print(cohort_pivot)

# Преобразуем в проценты (retention rate)
cohort_sizes = cohort_pivot[0]
retention = cohort_pivot.divide(cohort_sizes, axis=0) * 100

# Визуализация тепловой картой
plt.figure(figsize=(12, 8))
sns.heatmap(retention, annot=True, fmt='.1f', cmap='Blues')
plt.title('Retention rate по месяцам')
plt.xlabel('Месяц после первой покупки')
plt.ylabel('Когорта (месяц первой покупки)')
plt.show()
```
